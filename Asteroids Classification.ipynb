```python
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn import tree
from sklearn.cluster import KMeans
from sklearn.ensemble import RandomForestClassifier
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv ("Downloads/nasa-asteroids-classification/nasa.csv")
```

```python
del data ['Est Dia in M(min)']
del data ['Est Dia in M(max)']
del data ['Est Dia in Miles(min)']
del data ['Est Dia in Miles(max)']
del data ['Est Dia in Feet(min)']
del data ['Est Dia in Feet(max)']
del data ['Relative Velocity km per hr']
del data ['Miles per hour']
del data ['Miss Dist.(lunar)']
del data ['Miss Dist.(kilometers)']
del data ['Miss Dist.(miles)']
del data ['Orbiting Body']
del data ['Equinox']
del data ['Orbit Determination Date']
del data ['Close Approach Date']
data.head
```

```python
data.shape
```

```python
features = data.iloc [:, 0:24]
labels = data.iloc [:, [24]]
```

```python
scaler = MinMaxScaler()
scaler.fit (features)
features_transform = scaler.transform (features)
print (features_transform)
```

```python
pca = PCA (n_components = 20)
principle_components = pca.fit_transform (features_transform)
#principle_components = pca.fit_transform (features)
principleDf = pd.DataFrame(data = principle_components)
```

```python
labels
labels = np.asarray(labels)
print (labels)
```

```python
X_train, X_test, y_train, y_test = train_test_split (features_transform, labels)
```

```python
data.isnull().values.any()
```

```python
model = SVC()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict (X_test)
```

```python
accuracy = accuracy_score (y_pred, y_test)
print (accuracy)
```

```python
LRModel = LogisticRegression().fit(X_train, y_train)
LRPredict = LRModel.predict(X_test)
```

```python
accuracy = LRModel.score (X_test, y_test)

accuracy = accuracy_score (LRPredict, y_test)
print (accuracy)
```

```python
treeModel = tree.DecisionTreeClassifier().fit (X_train, y_train)
```

```python
treePredict = treeModel.predict (X_test)
```

```python
accuracy = accuracy_score (treePredict, y_test)
print (accuracy)
```

```python
RFModel = RandomForestClassifier(random_state = 12).fit (X_train, y_train)
RFPredict = RFModel.predict(X_test)
```

```python
accuracy = accuracy_score (RFPredict, y_test)
accuracy
```
